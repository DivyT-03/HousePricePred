# House Rent Prediction Model  Using Python

The House Price Dataset is taken from Kaggle : House Prices - Advanced Regression Techniques<br>
Link: https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data

```python
import pandas as pd
import numpy as np
```

## Data Preprocessing

```python
#Reading DataSet
df=pd.read_csv('C:\\Users\\inasu\\Downloads\\House\\train.csv')
```

```python
df.head()
```

```python
print("The dataset is of shape",df.shape)
```

We drop all the Columns with more than 500 Null Values. As they won't add much value to our prediction. 

```python
df=df.drop([df.columns[j] for j,i in enumerate(list(df.isnull().sum())) if i>500],axis=1)
```

We fill in the Null values with the mean values of the Int/Float Columns

```python
df.fillna(df.mean(),inplace=True)
```

```python
df.head()
```

## Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Heatmap to observe the coorelation between the colums

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

We drop the columns with a corr less than 0.3 as they are weak

```python
df=df.drop([df.columns[j] for j,i in enumerate(list(df.corr()["SalePrice"])) if i<0.3 and i>-0.3],axis=1)
```

```python
df.head()
```

For the categorical values, we use pandas get_dummies to fill in the dummy values.

```python
df_dummies=pd.get_dummies(df)
```

```python
df_dummies.head()
```

## Prediction 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
cols=list(df_dummies.columns)
cols.remove('SalePrice')
X_train, X_test, y_train, y_test = train_test_split(df_dummies[cols], df_dummies['SalePrice'], test_size=0.33, random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
r_sq = model.score(df_dummies[cols],df_dummies['SalePrice'])
print("Score of the model is:",r_sq,"with the intercept:",model.intercept_,"and coefficients:\n",model.coef_)
intercept=model.intercept_
coef=model.coef_
```

```python
y_hat = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score,jaccard_score,recall_score,accuracy_score,r2_score
from sklearn import metrics
RSS=np.mean((y_test-y_hat)**2)
VS=model.score(X_test,y_test)
MAE=np.mean(np.absolute(y_test-y_hat))
R2=r2_score(y_test,y_hat)
MSE=np.sqrt(metrics.mean_squared_error(y_test,y_hat))
```

```python
print("RSS=",RSS,"\nVS=",VS,"\nMAE=",MAE,"\nR2=",R2,"\nMSE=",MSE)
```
